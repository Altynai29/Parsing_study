# HTML Page layout


**HTML** (HyperText Markup Language) – это стандартный язык разметки, используемый для создания веб-страниц и веб-приложений. HTML описывает структуру веб-страницы и использует так называемые "теги" для обозначения различных элементов страницы, таких как текст, ссылки, изображения и другие.




## Основные Компоненты HTML
1. **Теги**: Основные строительные блоки HTML. Теги определяют, как браузер должен отображать содержимое. Примеры тегов: `<p>` для абзацев, `<a>` для ссылок, `<img>` для изображений.

2. **Элементы**: Элементы HTML состоят из открывающего тега, содержимого и закрывающего тега. Например, `<p> Это абзац.</p>`. Некоторые элементы, такие как `<img>`, не требуют закрывающего тега и называются "пустыми элементами".

3. **Атрибуты**: Атрибуты предоставляют дополнительную информацию о HTML-элементе, такую как источник изображения в `<img src="image.jpg">` или адрес ссылки в `<a href="https://example.com">`.

## Основные Теги HTML

* `<!DOCTYPE html>`: Определяет тип документа – HTML5.
* `<html>`: Корневой элемент, который обрамляет всю HTML-страницу.
* `<head>`: Содержит метаинформацию о веб-странице, такую как ее заголовок (`<title>`), ссылки на стилевые файлы и скрипты.
* `<body>`: Содержит весь контент веб-страницы, такой как текст, изображения, видео, ссылки и т.д.

## Распространенные Теги и Их Назначение

* `<h1>`,`<h2>`, ..., `<h6>`: Заголовки разного уровня.
* `<p>`: Абзац текста.
* `<a>`: Гиперссылка.
* `<img>`: Изображение.
* `<ul>`, `<ol>`, `<li>`: Неупорядоченные (`<ul>`) и упорядоченные (`<ol>`) списки с элементами списка `<li>`.
* `<table>`, `<tr>`, `<td>`: Таблица с рядами `<tr>` и ячейками `<td>`.
* `<div>`: Блоковый элемент для группировки контента и стилевого оформления.
* `<span>`: Строчный элемент для группировки контента внутри других элементов.
* `<form>`, `<input>`, `<button>`: Формы и элементы управления для сбора данных от пользователя.


## Расширенные Элементы и Функциональность
* **Формы и элементы ввода:** для создания интерфейсов, позволяющих пользователю отправлять данные на сервер.
* **Интерактивные элементы:** такие как `<button>`, `<select>`, `<textarea>`.
* **Семантические элементы:** такие как `<article>`, `<section>`, `<header>`, `<footer>`, `<nav>`, которые помогают создавать структурно организованный и доступный контент.


**HTML** – это не статический язык; он постоянно развивается. HTML5 – это последняя версия, включающая новые элементы, атрибуты и поведение, а также более широкую поддержку мультимедиа и графики.

# BeautifulSoup

Когда вы используете `BeautifulSoup(response.text, "html.parser")` в Python с использованием библиотеки **BeautifulSoup**, это создает объект `BeautifulSoup`, который представляет собой разобранное (проанализированное) дерево HTML-документа. Этот объект позволяет вам навигировать по структуре HTML-документа, извлекать нужные данные и манипулировать HTML-кодом.

Теперь, имея объект `soup`, вы можете использовать различные методы **BeautifulSoup** для поиска, извлечения и манипуляции данными в HTML.

# Methods `find()` & `find_all()`

`find()` и `find_all()` являются методами, используемыми в библиотеке **BeautifulSoup** для Python, которые помогают извлекать данные из HTML и XML-документов. Они позволяют искать определенные теги и атрибуты внутри документа, облегчая процесс веб-скрапинга.


## `find()`

Метод `find()` используется для поиска первого элемента, который соответствует указанным критериям. Если элементы, соответствующие критериям, существуют в документе, `find()` возвращает первый из них. Если таких элементов нет, возвращается `None`.

Пример использования `find()`:





```python
soup = BeautifulSoup(html_doc_1, 'html.parser')
title_tag = soup.find('title')
title_tag
```

Атрибут `.text` этого объекта позволяет получить только текстовое содержимое тега, игнорируя все HTML-теги внутри него. Это полезно, когда вам нужно работать только с текстом, а не с разметкой.

```python
title_tag.text
```

```python
soup.find('a')
```

```python
soup.find('a')['href']
```

```python
soup.find_all('p')
```

## `find_all()`

Метод `find_all()`, с другой стороны, возвращает список всех элементов, соответствующих указанным критериям. Если соответствующих элементов нет, возвращается пустой список.

Пример использования `find_all()`:

```python
soup = BeautifulSoup(html_doc_1, 'html.parser')
all_paragraphs = soup.find_all('p')
all_paragraphs
```

```python
for i in all_paragraphs:
  print(i.text)
```

# Более сложный вариант поиска

```python
html_doc = """
<!DOCTYPE html>
<html>
<head>
    <title>Пример страницы</title>
</head>
<body>

    <div class="header">
        <h1>Заголовок страницы</h1>
    </div>

    <div class="content">
        <p>Это основной контент страницы.</p>
        <div class="inner-content">
            <p>Это вложенный контент внутри основного контента.</p>
        </div>
    </div>

    <div class="footer">
        <p>Футер страницы</p>
    </div>

</body>
</html>
"""
```

В этом примере:

* Есть три основных `<div>` элемента с классами header, content и footer.
* Внутри `<div class="content">` есть вложенный `<div>` с классом inner-content.

Вы можете использовать BeautifulSoup для поиска этих элементов. Вот пример того, как можно использовать find() для поиска первого `<div>` с классом *content*:

```python
soup = BeautifulSoup(html_doc, 'html.parser')
content_div = soup.find('div', class_='content')
content_div
```

Поиск вложенного элемента:

```python
content_div.find('div', class_ = 'inner-content').text.strip()
```

Поиск вложенного элемента напрямую:

```python
soup.find('div', class_='inner-content').text.strip()
```
